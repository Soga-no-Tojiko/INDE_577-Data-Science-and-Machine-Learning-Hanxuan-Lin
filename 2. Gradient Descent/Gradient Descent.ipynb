```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_theme()
```

```python
# To illustrate how to conduct gradient descent, we will first generate a function that has multiple local minima.

def yf(x):
    return(np.sin(x) + np.cos(np.sqrt(2) * x) + np.sin(np.sqrt(3) * x))

x = np.linspace(-8, 8, 5000)
y = yf(x)

plt.figure(figsize=[12,8])
plt.plot(x, y, label = '$f(x) = sin(x) + cos(√2*x) + sin(√3*x)$')
plt.scatter([0.5], [yf(0.5)], color = "magenta", label = "$w_0$: guess no.1")
plt.scatter([4.5], [yf(4.5)], color = "red", label = "$w_0$: guess no.2")
plt.legend(fontsize = 12, loc = 'upper left')
plt.show()
```

```python
# If we luckily choose a starting point near the global minima(around -1.75)

x_start = 0.5
precision = 0.0001
learning_rate = 0.1
```

```python
def dyf(x):
    return( np.cos(x) - np.sqrt(2)*np.sin(np.sqrt(2)*x) + np.sqrt(3)*np.cos(np.sqrt(3)*x) )

def derivative_descent(derivative_function, alpha = 0.8, w_0 = 5.0, max_iter = 1_000):
    W = [w_0]
    i = 0
    while abs(derivative_function(W[-1])) > 0.001 and i < max_iter:
        w_new = W[-1] - alpha*derivative_function(W[-1])
        W.append(w_new)
        i += 1
    W = np.array(W)

    return W
```

```python
W = derivative_descent(dyf, alpha=learning_rate, w_0=x_start)

for i, w in enumerate(W):
    print(f"w_{i} = {np.round(w, decimals = 2)} | df(w_{i}) = {dyf(w)}")
print("")
```

### So we find a local minima at x = -1.44.

```python
# Figure of the first local minima

plt.figure(figsize=[12,8])
plt.plot(x, y, label = '$f(x) = sin(x) + cos(√2*x) + sin(√3*x)$')
plt.scatter([0.5], [yf(0.5)], color = "magenta", label = "$w_0$: guess no.1")
plt.scatter([4.5], [yf(4.5)], color = "red", label = "$w_0$: guess no.2")
plt.scatter(W, yf(W), color = "blue")
plt.scatter(W[-1], yf(W[-1]), color = 'yellow')
plt.legend(fontsize = 12, loc = 'upper left')
plt.show()
```

```python
# Or if we choose a starting point near another local minima(around 6.25)

x_start_2 = 4.5
```

```python
W = derivative_descent(dyf, alpha=learning_rate, w_0=x_start_2)

for i, w in enumerate(W):
    print(f"w_{i} = {np.round(w, decimals = 2)} | df(w_{i}) = {dyf(w)}")
print("")
```

### This time the function converges at another local minima of x = 6.27.

```python
# Figure of the second local minima

plt.figure(figsize=[12,8])
plt.plot(x, y, label = '$f(x) = sin(x) + cos(√2*x) + sin(√3*x)$')
plt.scatter([0.5], [yf(0.5)], color = "magenta", label = "$w_0$: guess no.1")
plt.scatter([4.5], [yf(4.5)], color = "red", label = "$w_0$: guess no.2")
plt.scatter(W, yf(W), color = "blue")
plt.scatter(W[-1], yf(W[-1]), color = 'yellow')
plt.legend(fontsize = 12, loc = 'upper left')
plt.show()
```
